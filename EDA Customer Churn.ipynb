```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")

df = pd.read_csv("Telco_Customer_Churn_Clean.csv")
```

```python
#Overall churn
plt.figure(figsize=(6,4))
sns.countplot(data=df, x="Churn")
plt.title("Cusomer Churn Distribution")
plt.xlabel("Churn")
plt.ylabel("Numbers of Customers")
```

```python
#Business Question: What percentage of customers churned?
#Observation: More customers stayed than left.
#Business Insight: The company retains most customers, but the number of churned customers is significant enough to investigate."
```

```python
#Main churn driver
plt.figure(figsize=(8,5))

sns.countplot(
    data=df,
    x="Contract",
    hue="Churn"
)

plt.title("Churn by Contract Type")
plt.show()
```

```python
#Business Question: Which Contract has the highest churn?
#Observation: Month-to-month customers have the highest churn.
```

```python
#Customer behavior
plt.figure(figsize=(10,5))
sns.countplot(
    data=df,
    x="PaymentMethod",
    hue="Churn"
)
plt.xticks(rotation=20)
plt.title("Customers Churn by Payment Method")
plt.show()
```

```python
#Business Question: Does payment method affect churn rate?
#Observation: Electronic Check payment method loses the most.
```

```python
#Pricing analysis
plt.figure(figsize=(7,5))
sns.boxplot(
    data=df,
    x="Churn",
    y="MonthlyCharges"
)
plt.title("Churn by Monthly Charges")
plt.show()
```

```python
#Business Question: Do higher monthly bills increase churn rate?
#Observation: High monthly charges affect the churn rate.
```

```python
#Customer lifecycle
plt.figure(figsize=(8,5))
sns.histplot(
    data=df,
    x="tenure",
    hue="Churn",
    bins=30,
    kde="True"
)
plt.title("Customer Tenure Distribution")
plt.show()
```

```python
#Business Question: Do new customers tend to leave early?
#Observation: New customers leave earlier.
```

```python
#Service Analysis
plt.figure(figsize=(7,5))
sns.countplot(
    data=df,
    x="InternetService",
    hue="Churn"
)
plt.title("Churn by Internet Services")
plt.show()
```

```python
#Business Question: Does internet service influence churn rate?
#Observation: Fiber Optic Internet service has higher churn.
```

```python
numeric_df = df.select_dtypes(include=['int64','float64'])

plt.figure(figsize=(8,6))
sns.heatmap(
    numeric_df.corr(),
    annot=True,
    cmap="coolwarm"
)
plt.title("Correlation Heatmap")
plt.show()
```
